```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/DataMining/Data01.csv')
df.head(8)
print(df.info())
```

```python
print(df.shape)
print(df.dtypes)
print(df.columns)

colNames = df.columns.tolist()
print(colNames)

print(len(df))

df.describe()
print(df['출고수량'].describe())
```

```python
df['카테고리'].value_counts()
```

```python
total_column = df.columns.tolist()

num_column = df.describe().columns.tolist()

char_column = []
for i in total_column :
  if i not in num_column :
    char_column.append(i)

df[char_column].describe()
```

```python
df.sort_values(by='입고수량', ascending=False)
df.sort_values(by=['입고수량','출고수량'], ascending = [True,False])
```

```python
df.info()
```

```python
df['날짜']
df['날짜2'] = pd.to_datetime(df['날짜'])
print(df['날짜2']) ##원본데이터 유지하기위해
```

```python
df['년'] = df['날짜2'].dt.year
df['월'] = df['날짜2'].dt.month
df['일'] = df['날짜2'].dt.day
df['주차'] = df['날짜2'].dt.week
df['요일'] = df['날짜2'].dt.day_name()
df
```

```python
pd.pivot_table(data = df, index = '월', values='출고수량', aggfunc='sum').plot(kind='bar')
```
